```python
import pandas as pd
import numpy as np

import boto3
import sagemaker 
```

```python
sagemaker_session = sagemaker.Session()

bucket = 'cjai-raw-initial-s3'
prefix = 'test/dsc-workshop/train'
role = sagemaker.get_execution_role()
```

```python
inputs = 's3://{}/{}/anomaly_dummy.csv'.format(bucket,prefix)
```

```python
from sagemaker.pytorch import PyTorch
```

```python
s3_output_location = 's3://{}/{}'.format(bucket,prefix)
estimator = PyTorch(entry_point='train_pt.py',
                    role=role,
                    framework_version='1.4.0',
                    train_instance_count=1,
                    train_instance_type='ml.m4.2xlarge',
                    output_path=s3_output_location,
                    hyperparameters={
                        'epochs': 2,
                        'backend': 'gloo',
                        'batch-size': 64,
                        'test-batch-size': 64,
                    })
estimator.fit({'training': inputs})
```
